```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd 
```

## Reading dataset

```python
dataset = pd.read_excel('perfume_data.xlsx',header=None)
dataset = dataset.transpose()

header = dataset.iloc[0]
dataset = dataset[1:]
dataset = dataset.rename(columns=header)

dataset = dataset.iloc[:2,:].transpose()
dataset
```

## KMeans

```python
kmeans = KMeans(n_clusters=3,random_state=0,init='random').fit(dataset)
```
